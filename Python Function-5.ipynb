### 2.  Arbitrary Arguments 

****If the number of arguments unknown then it need to simply add sterik ( * ) before the parameter****

****like *name****

****This way, the function will recieve a a tuple of arguments****

```python
def intro(*name):
    print('Welcome to the My repository! ', name[0], ', ', name[1], 'and ', name[2])
```

```python
intro('Wasim', 'Fahim', 'jhon')
```

****For any number of arguments?****

****loops can be used.****

##### Using while Loop

```python
def intro(*name):
    i = 0
    print('Welcome to Imtiaz\'s repository ', end = '\n')
    while len(name) > i:
        print(name[i], end = ', ')
        i += 1
```

```python
intro('mr.x','Mr.y', 'Mr.z')
```

### 3. Keyword Arguments (*kwargs*)

****In keyword arguments we explicitly specify a keyword to a certain value****

* ****keyword = value****

* ****While passing values to a function using function call, the keyword used must match the keyword that was assigned during function definition****

* ****kwargs are declared by a name and value****

```python
def fruit(quantity,items, price):
    print(quantity, items, 'cost', price,'Rs')
```

```python
fruit('apple', 900, 5)
```

```python
fruit(price = 1000, quantity = 6,items = 'banana' )
```

****The order of the arguments doesn't matter here but the keywords must match****

****What if we add another keyword while calling a function?****

### 4. Arbitary Keyword Arguments (***kwargs*)

****If the number of to be passed are unkown, then we can simply add two steriks ( ** ) before the parameter name****


****Here the function will recieve a dictionary of arguments****

```python
def fav_actors(**actors):
    for i in actors:
        print("His favourite bollywood actor is " + actors[i])
```

```python
fav_actors(actor1 = "akshykumar", actor2 = 'shahruh khan',actor3 = "salman khan", actor4 = "amita bachan")
```

### 5. Default Parameter Value

*Value assigned to a paremeter in a function definition is known as* **default value** *or optional value****

*If function is called without changing the paraameter then it will return its default value*

*default value can be changed while calling the function*

```python
def addition(x =0,y=0):
    print('The value of x: ', x)
    print('The value of y: ', y)
    return x+y
```

* **In the above function definition, two parameter with default value of 0s are give.**
* **If the function is called without passing any arguments, the default value will be given automaticaly**

```python
c = addition(10)
print(c)
d = addition(10,5)
print(d)
```

### $\text{Mutable Data Types as parameters}$

* **We can also add mutable data types as parameters of the function like lists, and dictionaries**
* **Mutable data types can also be added as parameters of the function, like list and dictionary.**

```python
colors = ['Red', 'Blue', 'Green', 'Orange', 'White']
def paint(colors):
    print('The colors are: ', colors)
```

```python
paint(colors)
```

```python
colors = ['Red', 'Purple', 'Blue', 'Green', 'Orange', 'White']
def painting(colors):
    for i in colors:
        print('The colors are: ', i)
```

```python
painting(colors)
```

```python
def paint(colors):
    print('Before appending: ', colors)
    new_color = input('Enter a color: ')
    colors.append(new_color)
    print('After appending: ', colors)
```

```python
painted(colors)
```
